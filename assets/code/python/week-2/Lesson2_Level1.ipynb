### Arranging Shapes in Specific Patterns Using Loops
* By using a for loop, we can instruct the computer to draw shapes automatically.
* Practice drawing circles and positioning them evenly along a horizontal line.
--- 

```python
fig, ax = sm.setup_figure('Arranging Shape in Line');

# Draw circle in line using for loop
for i in range(0,11,1):
    circle = patches.Circle((i, 2.5), 0.5, facecolor='blue')
    ax.add_patch(circle)
```
