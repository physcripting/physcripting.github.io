### Arranging Shapes with Two for Loops ðŸŽ¨ 
By using **for loops**, we can tell the computer to draw shapes automatically in rows and columns.
* The outer loop controls one direction (like rows)
* The inner loop controls another direction (like columns)
* This lets us create grids, patterns, and complex designs easily without drawing each shape by hand.

```python
fig, ax = sm.setup_figure('Arranging Shape in two dimensions');

for i in range(0,11,1):
    for j in range(0,6,1):
        circle = patches.Circle((i, j), 0.5, facecolor='blue')
        ax.add_patch(circle)
```
